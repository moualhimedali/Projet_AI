```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
! pip install simplet5 -q
```

```python
data = pd.read_csv("/kaggle/input/users-inputs-dataset/users_inputs_dataset.csv", index_col=0)
```

```python
data.head()
```

```python
data['source_text'] = "transform_text: " + data['source_text']
```

```python
data.head()
```

```python
train_data, test_data = train_test_split(data, test_size=0.2)
train_data, eval_data = train_test_split(train_data, test_size=0.1)
train_data.shape, eval_data.shape, test_data.shape
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.train(train_df=train_data,
            eval_df=test_data,
            source_max_token_len=128,
            target_max_token_len=50,
            batch_size=8, max_epochs=5, use_gpu=True)
```

```python
! ( cd outputs; ls )
```

```python
# let's load the trained model from the local output folder for inferencing:
model.load_model("t5","/kaggle/working/outputs/simplet5-epoch-4-train-loss-0.0086-val-loss-0.0179", use_gpu=True)
```

```python
user_input = "What is Monte Carlo analysis used for"
```

```python
model.predict(user_input)
```
